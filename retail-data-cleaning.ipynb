# Removing Negative Values from the "Quantity" Column

```python
df_no_negative_values = df[df['Quantity'] >= 0]
rows_removed = len(df) - len(df_no_negative_values)

print(f"Original dataframe: {len(df)} rows")
print(f"After removing negative quantities: {len(df_no_negative_values)} rows")
print(f"Number of rows removed: {rows_removed}")
```

# Replacing Null Values

#### Replacing Null Values in the "CustomerID" column

```python
int(df_no_negative_values['CustomerID'].isna().sum())
```

```python
# Replace null values in the "CustomerID" column with "Unknown Customer"
df_no_negative_values['CustomerID'].fillna('Unknown Customer', inplace=True)
```

```python
int(df_no_negative_values['CustomerID'].isna().sum())
```

#### Replacing Null Values in the "Description" Column

```python
int(df_no_negative_values['Description'].isna().sum())
```

```python
# Replace null values in the "Description" column with "Unknown Product"
df_no_negative_values['Description'].fillna('Unknown Product', inplace=True)
```

```python
int(df_no_negative_values['Description'].isna().sum())
```

# Renaming the "Description" column to "Product Name"

```python
df_no_negative_values = df_no_negative_values.rename(columns={'Description': 'Product Name'})
df_no_negative_values
```

# Adding a "Season" Column 

```python
# First, ensure InvoiceDate is in datetime format
df_no_negative_values['InvoiceDate'] = pd.to_datetime(df_no_negative_values['InvoiceDate'])

# Extract month from InvoiceDate
df_no_negative_values['Month'] = df_no_negative_values['InvoiceDate'].dt.month

# Create a mapping dictionary for months to seasons
season_mapping = {
    1: 'Winter',  # January
    2: 'Winter',  # February
    3: 'Spring',  # March
    4: 'Spring',  # April
    5: 'Spring',  # May
    6: 'Summer',  # June
    7: 'Summer',  # July
    8: 'Summer',  # August
    9: 'Fall',    # September
    10: 'Fall',   # October
    11: 'Fall',   # November
    12: 'Winter'  # December
}
```

```python
# Map months to seasons
df_no_negative_values['Season'] = df_no_negative_values['Month'].map(season_mapping)
```

# Top 10 Most Purchased Products 

```python
# Returns a series with the counts of each unique value in the 'StockCode' column
duplicate_counts = df['StockCode'].value_counts() 

# Returns a series with only the counts of the products that purchased more than once
duplicates = duplicate_counts[duplicate_counts > 1] 

print(f"Number of StockCode values that appear multiple times: {len(duplicates)}")
print("\nTop 10 most frequently occurring StockCodes:")
print(duplicates.head(10))
```

```python
top_stock_codes = duplicates.head(10).index # Gets all the StockCode unique identifiers
stockcode_descriptions = {}

# Updates the stockcode_descriptions dictionary with the "StockCode" being the key and "Description" being the value
for code in top_stock_codes:
    description = df.loc[df['StockCode'] == code, 'Description'].iloc[0]
    stockcode_descriptions[code] = description

# Prints the result by looping through each key-value pair in the dictionary
print("Top 10 most frequently occurring StockCodes:")
for code, description in stockcode_descriptions.items():
    print(f"StockCode: {code}, Description: {description}, Number of Purchases: {duplicates[code]}\n")
    
```
